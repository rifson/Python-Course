8) Create a new column called price-range, loop through the dataset:
 

*   Price below 10000 ==> price-range = cheap
*   Price between 10000 and 20000 (both included) ==> price-range = affordable
*   Price over 20000 ==> price-range = expensive 






```python
for i in range(len(auto_df)):
  if (auto_df.loc[i, 'price'] < 10000):
    auto_df.loc[i, 'price-range'] = 'cheap'
  elif ((auto_df.loc[i, 'price'] >= 10000) & (auto_df.loc[i, 'price'] <= 20000)):
    auto_df.loc[i, 'price-range'] = 'affordable'
  elif (auto_df.loc[i, 'price'] > 20000):
    auto_df.loc[i, 'price-range'] = 'expensive'

auto_df
```
