## **Logical Operators**

```python
#Logical and
a = True
b = False
c = a and b
print(c)
```

```python
#Logical or
a = True
b = False
c = a or b
print(c)
```

```python
#Logical not
a = True
b = False
c = not (a and b)
d = not (a or b)
print(c)
print(d)
```

```python
a = 11
b = 10
print(a<b)
print(a>b)
print(a<b and a>b)
```

```python
#Logical flow control operations
#find the largest of 2 numbers
a = 3
b = 3
if a > b:
  #do something here
  print("a is greater than or equal to b")
elif b > a:
  #do something here
  print("b is greater than a")
else:
  print("b is the same as a")
```

```python
#Check if a number is odd or even
a = 4
if a % 2 == 0:
  print("a is even")
else:
  print("a is odd")
```

# python data structures - Lists, Tuples, Sets, Dictionaries

A list is a collection of items in a particular order. You can make a list that
includes the letters of the alphabet, the digits from 09, or the names of all
the people in your family. You can put anything you want into a list, and the
items in your list dont have to be related in any particular way.

```python
list_nums = [3,4,7,23,1,4,23,90,0.5]
print(list_nums)
```

```python
cars = ['BMW', 'Ferrari', 'Rolls royce', 'Ford']
print(cars)
#Python considers the first item in a list to be at position 0, not position 1.
print(cars[0])
```

```python
# List numbers
list_nums = [3,4,7,23,1,4,23,90,0.5]
list_nums[5] #6th element
```

# Modifying list elements

```python
cars = ['BMW', 'Ferrari', 'Rolls royce', 'Ford']
print(cars)
cars[0] = 'Mclaren'
print(cars)
```

Adding elements

```python
cars = ['BMW', 'Ferrari', 'Rolls royce', 'Ford']
print(cars)
cars.append('Audi')
print(cars)
```

Inserting elements

```python
cars.insert(2, 'BMW')
print(cars)
```

Removing elements

```python
del cars[0]
print(cars)
```

# Dictionaries

Create dictionary

```python
simple_dict = {'Denmark' : 'Copenhagen'}
print(simple_dict)
```

```python
dict2 = {'Denmark' : 'Copenhagen', 'India' : 'New Delhi'}
print(dict2)
```

Adding new values

```python
dict4 = {} #Empty dictionary
dict4['Denmark'] = 'Copenhagen'
dict4['India'] = 'New Delhi'
print(dict4)
```

# Tuples

```python
first_tuple = (1,10,30)
first_tuple[1]
```

```python
my_tuple = (1, 2, 3)
#my_tuple.append(4)  # This would raise an error
#my_tuple[1] = 5     # This would also raise an error
my_tuple[1]
```

# Sets

Sets allow items to be stored unordered, unchangeable without any duplicates

```python
cars = {"BMW", "Audi", "Benz"}
print(cars)
```

```python
cars = {"BMW", "Audi", "Benz","Audi"} #multiple values are ignored
print(cars)
```

```python
my_set = {1, 2, 3}
my_set.add(4)    # Add an element
my_set.remove(2)  # Remove an element

print(my_set)  # Output: {1, 3, 4} (order might vary)
```
