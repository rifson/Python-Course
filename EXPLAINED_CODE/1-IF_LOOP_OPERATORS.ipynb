# Introducing Variables

What are variables?

Create variables with suitable names
Store values into the variables
retrieve the variables as needed

In other programming languages, such as C++,C#, variables must be rigorously defined by type. That is an integer variable can only hold whole numbers and string variables can only hold text.

But a python variable can hold anything. In other words - universal

Python variables

Variable names may contain  digits (0-9) , letters or the underscore character _ .
Variable names must begin the underscore _ character or with a letter from A-Z
Variable names may not be a reserved word in Python



```python
one = 1
print (one)
```

```python
great_you = 'hi'
print (great_you)
```

## Assignment Operators

Assign: Operator is used to assign the value of the right side of the expression to the left side operand.

```python

variable1 = 3
variable2 = 5

variable3 = variable1 + variable2

# Output
print(variable3)
```

```python
a = 300
b = 10

# Add and Assign",
a += b
print("Add and Assign:",a)

#Subtract and Assign:
a -= b
print("Subtract and Assign:",a)

#"Multiply and Assign",
a *= b
print("Multiply and Assign:",a)

# "Divide and Assign:"
a /= b
print("Divide and Assign:",a)



```
