## **URL to the webpage "2018-19 NBA Player Stats: Per Game"**

### Method 1

We could modularize the URL and year strings of text that may be useful for programmatic access (in the subsequent code cell).

```
# The Building Blocks
year = '2019'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

# Combining the URL + year strings together
url = url_link.format(year)
url

#The format(year) method is used to insert the value of the year variable into the url_link string. In this specific case, you have a placeholder {} within the url_link string, and format(year) replaces this placeholder with the actual value of the year variable ('2019').
#Placeholder: The {} within the string acts as a placeholder where you want to insert a dynamic value.
#Format() Method: The format() method is called on the string (url_link), and it takes the year variable as an argument.
#Replacement: The format() method replaces the {} placeholder with the value of the year variable, resulting in the final formatted string.
#Why use this approach?
#Flexibility: This makes your code more dynamic, as you can easily change the year variable to fetch data for different years without having to manually modify the entire URL string.
#Readability: It improves the readability of your code, as the URL structure is clearly defined with a placeholder, making it easier to understand the intention.
```

Now, let's programmatically return a list of URL given a list of years (e.g. 2015,2016,2017,2018,2019).

```
years = [2015,2016,2017,2018,2019]
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

for year in years:
  url = url_link.format(year)
  print(url)
```

## **Read HTML webpage into pandas**

```
import pandas as pd
```

### Using pd.read_html() function to read HTML webpage

```
df = pd.read_html(url, header = 0)
df
#In this case, header = 0 indicates that the first row of the HTML table should be used as the column names for the resulting DataFrame. By default, if the header parameter is not specified, read_html() will attempt to infer the column names from the HTML table.

#Setting header = 0 explicitly ensures that the first row of the HTML table is treated as the header row and not as part of the data in the DataFrame.
```
